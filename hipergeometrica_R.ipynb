## Distribución Hipergeometrica

Probalidad de encontrar 4 ases es un juego de 5 cartas

```r
library(tidyverse)
```

función para calcular la distribución hipergeometrica

probabilidad de la jugada Color

```r
color <- dhyper(5, 13, 47, 5) * 100
color
```

probabilidad de la jugada Trío

```r
trio <- dhyper(3, 5, 47, 5) * 100
trio
```

probabilidad de la jugada poker

```r
poker <- dhyper(4, 5, 47, 5) * 100
poker
```

comparación de probabilidades

probabilidad de encontrar desde 0 a 4 ases en una mano

```r
df <- data.frame(As = c(0, 1, 2, 3, 4),
                   Probabilidad = c(dhyper(0:4,5,47,5)*100))
df
```

```r
plot(df,type= "b", xlab="Ases",ylab="Probabilidad",main="Función de distribución Hipergeometrica", col = 'red')
```

```r
df%>%
     ggplot(., aes(As,Probabilidad))+
     geom_col(color = "darkred", fill = "lightblue")+
     theme_bw()+
     labs(x="N de Ases", y="Probabilidad")+
     theme(axis.title = "element_text"(size = 18),
        legend.title = element_text(size = 14), 
        legend.text = element_text(size = 14))
```

Qué pasa en una baraja española

```r
df2 <- data.frame(As = c(0, 1, 2, 3, 4),
                   Probabilidad = c(dhyper(0:4,5,35,5)*100))
df2
```

```r
df2%>%
     ggplot(., aes(As,Probabilidad))+
     geom_col(color = "darkred", fill = "lightblue")+
     theme_bw()+
     labs(x="N de Ases", y="Probabilidad")+
     theme(axis.title = "element_text"(size = 18),
        legend.title = element_text(size = 14), 
        legend.text = element_text(size = 14))
```

grafica de las dos distribuciones

```r
plot(df, type= "b", xlab="Ases",ylab="Probabilidad", main="Comparación de probabilidades", col = 'blue')
lines(df2, lty = 2, col = 'red')
legend("topright", legend=c("baraja inglesa", "baraja española"),
       col=c("blue", "red"), lty = 1:2, cex=0.8)
```
